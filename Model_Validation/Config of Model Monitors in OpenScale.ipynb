## Configuration for Model Monitors  in Watson OpenScale

#### Navigate to the OpenScale Dashboard, and click on the tile of the model. This will take you to a screen shown below. Then in the `Actions` dropdown, click on `Configure monitors`.


- `Configure monitors` will bring open a new page, like below.

### Now follow this wizard to provide rest of the information to configure the monitors.
- When it asks for feature and categorical feature selection, please select appropriate columns. 
- Select each item (Quality, Fairness, Drift) from the left side and configure using the information below.

#### Configuring Monitor for Explainability -
- The steps in previous notebook have already configured Explainability as you will see in the left Pane.

#### Configuring Monitor for Fairness -

- The steps in previous notebook have already configured Explainability as you will see in the left Pane.


#### Configuring Monitor for Quality -

- For Area under ROC threshold, use 0.8.

- Use 50 as minimum sample.


#### Configuring Monitor for Drift -

Description: In pre-production environments, the drift monitor computes the actual accuracy by using test data. For pre-production, the drift monitor reads the input data from feedback table, instead of the payload logging table, so that the label column is available for determining drift from the test data.

- Click on the `Pencil` icon next to the tile which says `Drift Model`
- Select the training option as `Train in a data science notebook`
- Upload the drift_detection_model.tar.gz that was downloaded in to your local system while working on `Create Drift Detection Model` Notebook 
- Click on `Save` as given the below screenshot


<br>

- Edit the `Drift Threshold` tile by click on the `Pencil` icon and set the default drift alert threshold as 10%

- Similarly, Edit the sample size to 50 as given in the screenshot below
