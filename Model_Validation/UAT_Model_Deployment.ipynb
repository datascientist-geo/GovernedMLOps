## Model Deployment in UAT/Pre-prod Deployment Space


#### This sub-step involves staging our model in UAT/Pre-prod deployment space and configure the deployed UAT model for monitoring performance in Openscale.  Run all cells below once in order to activate the links for the respective notebooks.

<font color = 'green'>*Note: UAT and Pre-prod can be used interchangeably*. </font>

### Pre-requisites

You should have a pre-existing UAT Deployment space with the name "Pre_Prod_Deployment_`GroupNumber`"


### Step 1. UAT Deployment Space 

Go to the Navigation menu of your Cloud Pak for Data Platform, and select `Deployments` to view all your deployment spaces.

Open the "Pre_Prod_Deployment_`GroupNumber`" deployment space

#### Go to `Access Control` of this deployment space and make sure that there is a user named `admin` in the Collaborators section. 
- This is needed for Openscale

<i><b>Note:</b>  If you are logged in as ‘admin’, then you would not have to check for ‘admin’ as a collaborator. </i>
 

<br>

### Step 2. Model Deployment in UAT Space

Follow the steps below to stage the model in your UAT Deployment Space. 
   
   - Go to the `Assets` tab of your 'Airlines-MLops-`GroupNumber` project. 
   - Scroll down to the list of the Models. 
   - Go to the model you want to publish to the "Pre_Prod_Deployment_`GroupNumber`"
   - Hover on the triple vertical dots at the right of the model name and click `Promote` as given below



If the "Pre_Prod_Deployment_`GroupNumber`" UAT space is associated to the project, You will be on the "Promote to Space" page. 

Make sure that the target space is "Pre_Prod_Deployment_`GroupNumber`". Click on `Promote` and the model will be promoted to that deployment space successfully.

<br>

<i> 
    
Note: If a space is not associated with the project yet, it would popup an option to associate a deployment space. 
    
- In the popup, select `Associate Deployment Space`. 
- In the next window, click on `Existing` tab
- Select the "Pre_Prod_Deployment_`GroupNumber`" UAT space then click `Associate`. 
- Now, promote the model again. This time the model should be promoted to the deployment space.
    
</i> 


Go to your "Pre_Prod_Deployment_`GroupNumber`" deployment space by selecting the `Deployments` in the Navigation menu.

To stage your desired model, click on the `Deploy` icon as shown in the image below.

<br>

You will now reach the "Create a Deployment" page.  

- Select `Online` as the deployment type
- Give a name to your model deployment, for example, "Model_Deployment_`GroupNumber`"
- Add additional description, if required
- Finally, click on the `Create` button.


<br>

You will now be back in your "Pre_Prod_Deployment_`GroupNumber`" deployment space

- Go to the `Deployments` tab where you will be able to see the status of your model deployment.
- The deployment status will change from `In progress` to `Deployed` once the model staging is successful completed.
      
<br>

### Step 3 - Payload Logging 

Open the "Model_Deployment_`GroupNumber`" model deployment. 

This will take you to a new page where you can send payload scoring request. You can send the initial payload scoring request to test the model here by specifying values of all different features/attributes in the screen.

- Select `Test` tab to send a payload scoring request. 
- Click on `Provide input data as JSON` button.
- Copy the payload below in the `Enter Input Data` section to begin the testing.

<code>
{
	"input_data":[{
	"fields": ["DAY","DAY_OF_WEEK","ORIGIN_AIRPORT","DESTINATION_AIRPORT","DEPARTURE_DELAY","TAXI_OUT","DISTANCE"],
	"values": [[2,5,"PHX","MSP",-2.0,13.0,983]]
	}]
}
</code>

- Press `Predict`.
