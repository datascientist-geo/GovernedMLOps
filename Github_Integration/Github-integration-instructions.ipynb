# Creating a project with Github integration

1. From Cloud Pak for Data platform window, click on the Projects. Then, click on the 'New project' button shown in below image.

2. Select <mark>Analytics project</mark> as the project type and click on Next button

3. Select Create an empyt project button as shown with red arrow below image

4. Fill in the define project details. First, enter name of the project. Second, in the choose project options section put a check mark in forn of <mark>Integrate this project with Git</mark> section. If you did not create a token yet, click on the <mark>New Token +</mark> section that was shown in the red rectangle below. If you already created a token, then please proceed to step 6.

5. On the Git integration window, select platform as Github or Github Enterprise (depending on the platform you would like to use). Then, enter your access token you copied from your Github page or Github Enterprise page. Next, enter the Domain URL. If you are using github, the correct url is <mark>https://github.com</mark> For the enterprise addition of github, please use the appropriate domain url. 

`Note: In case you do not have github token, please follow the instructions at this link` 
https://docs.github.com/en/github/authenticating-to-github/keeping-your-account-and-data-secure/creating-a-personal-access-token

Next, give your token a name and click on the continue button. 

6. Click on the Select Token cell and pick the token from your list. Enter the Github repository URL that you created for this project and select the master branch as master. Finally, put a check mark for <mark>Edit notebooks and Python scripts with the JupyterLab IDE</mark> section and click on the create button at the right bottom corner of the page.

The process should take some time. Once it is done, click on the view new project button to view your newly created project.
