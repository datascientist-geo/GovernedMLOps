## UAT Deployment Space Creation and Model Deployment


#### This sub-step involves creating a UAT environment for deploying our model and configure UAT model for monitoring performance in Openscale.  Run all cells below once in order to activate the links for the respective notebooks.

<font color = 'green'>*Note: UAT and Pre-prod can be used interchangeably*. </font>

### Step 1. UAT Deployment Space 

+ Go ahead and create a deployment space in your Cloud Pak for Data by following the steps below: 
 * If you have not already created a Deployment Space, go to left Navigation Menu (the top left hamburger icon) and click on the `Deployments` link.

 * There create a new Deployment Space by clicking the `New Deployment Space +` link.
 * Select `Create an empty space`. 
 * Name the deployment space as "UAT_Deployment_Space" or another desired name for UAT.

 * Click on `Create`.
 
Pictorial steps are given in the following cells.
 

#### Go to `Access Control` of this deployment space and add `admin` as collaborator by clicking on `Add collaborators` and typing admin in the pop-up box.
- This is needed for Openscale
- When it is done, `admin` should appear in the list of collaborators as shown in the picture below.

<i><b>Note:</b> If you are logged into the cluster as the 'admin' user, this step is not needed, as you are already a main collaborator of the deployment space. </i>

<br><br>

### Step 2. Model Deployment in UAT Space

Follow the steps below to deploy the model you created in the model development step in this UAT Deployment Space. 
   
   - Go to the `Assets` tab of the project. There, scroll down to the list of the Models. The Model you saved in previous step should appear there. 
   - Hover on the triple vertical dots at the right of the model name and click `Promote`. 
       - If a space is not associated with the project yet, it would popup an option to associate a space. In the popup, press `Associate Deployment Space`. In the next window, click on `Existing` tab and select the UAT space you created earlier, then click `Associate`. Now, promote the model again. This time the model should be promoted to the deployment space.
       - If a space is already associated to the project, the model will be promoted to that deployment space.



- Now go to the space ( left hamburger `menu`-> `Deployments`-> UAT_Deployment_Space) and click on the `Deploy` icon as shown below to deploy the desired model. 

<br>

   * That will bring a deployment screen. There, you select `Online`. <font color = 'red'> Note: Give a UNIQUE NAME of the deployment </font> and additional description. Then, click on the `Create` button.


<br>

   * In the next screen, you will see that the deployment is in progress and eventually, the Model will be shown in `Deployed` status.
      
<br>

### Step 3 - Payload Logging 

- Next, you can click on the deployment (from the deployment space, UAT_Deployment_Space) and it will take you to a new page where you can send payload scoring request. You can send the initial payload scoring request to test the model here by specifying values of all different features/attributes in the screen.
- Select `Test` tab to send a payload scoring request. 
- Use the payload below for testing.

<code>
{
	"input_data":[{
	"fields": ["DAY","DAY_OF_WEEK","ORIGIN_AIRPORT","DESTINATION_AIRPORT","DEPARTURE_DELAY","TAXI_OUT","DISTANCE"],
	"values": [[2,5,"PHX","MSP",-2.0,13.0,983]]
	}]
}
</code>

- Press `Predict`.
