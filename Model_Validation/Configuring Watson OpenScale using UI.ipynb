## Configuration of Watson OpenScale for CPD Cluster

#### This is an one time activity of configuring Watson Open Scale in your CPD cluster for Database and Machine Learning Provider.

#### In a new tab, please open Watson OpenScale homepage (`<hostname>/aiopenscale`). For example - https://zen-cpd-zen.apps.cpdv35-swat.cpd-daell.com/aiopenscale. 
- You will see an `Insights Dashboard` similar to the following picture. 

- Navigate to `Configure` tab on left pane.


##### Verify that a Datamart database is configured
   - If a database is already configured (green/blue checkmark beside `Database`), please skip to the next step.<font color = 'red'> DO NOT change the existing database credentials!</font>
   - If a database is not configured, please provide database access information by editing `Database` or contact your administrator. Please note that IBM Cloud Pak for Data currently only support DB2 as Datamart.

##### Add Machine Learning Provider
   - Select `Machine learning providers` from the left pane and click on `Add machine learning provider`.

<br>

- Now, on the next screen, provide a unique provider name. Then click on the edit icon of the `Connection`. 

<br>

- Then select `Watson Machine Learning (V2)` as the service provider type as shown in the picture below.
- Select the UAT space we created previously in the `Select space` dropdown. If you don't see your space here, please revisit `adding admin as collaborator` for deployment space in previous notebooks.  
- Make sure the environment type is selected as `Pre-production` in case you're deploying the model in UAT/Pre-prod environment. Then click on `Save`.
