# Publish Model to a Catalog

## Overview 


#### This notebook outlines the steps to publish an existing model to your respective catalog, Model_Catalog_`GroupNumber`.  This catalog has been pre-created on your Cloud Pak for Data Cluster.

#### Models are published to the catalog so that they can be retreived by other projects. Make sure to perform this step for every model in your Airlines-MLOPs-`GroupNumber` project. 

## Prerequisites

####  This notebook assumes you have the following assets:
- A pre-existing catalog named Model_Catalog_`GroupNumber`
- A model saved in your Airlines-MLOPs-`GroupNumber` project. 

Note: If you do not have a model, create one prior to proceeding with the below steps.


## View your Catalog

#### To view your Model_Catalog_`GroupNumber` catalog, go to the navigation menu on your Cloud Pak for Data Platform and click on `All Catalogs` under the 'Catalogs' section. 

This will take you the page where you will be able to view all your catalogs

<br>



<br>

You can now proceed on to publishing the model to Model_Catalog_`GroupNumber` catalog 

## Publish Model to Catalog

1. Navigate to the `Models` section in the project, Airlines-MLOPs-`GroupNumber`. This is where your `Airline_AutoAI_exp` model should be saved. 

- In this case, we are using "airline-scikitlearn" model as an example

2. Publish this model to  Model_Catalog_`GroupNumber` Catalog, by selecting `Publish to Catalog` from the drop-down options available for your choice of model

3. Select the name of your pre-existing Catalog, that is, Model_Catalog_`GroupNumber`. Set Privacy to `Public`, and then select `Publish`. 

- In this case, we are using "Model Assets" as our Target Catalog. 



4. Once the model has been published, it will appear in the Model_Catalog_`GroupNumber` Catalog.

- In this case, we are using "Model Assets" catalog as an example

5. Make sure to repeat this process for all the other models as well. You will be able to view your published model in the "Browse assets" section of your Model_Catalog_`GroupNumber` Catalog as shown in the image below
