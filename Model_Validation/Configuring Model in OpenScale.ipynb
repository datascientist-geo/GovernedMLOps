## Configuring Model in OpenScale


 #### This sub-step involves configuring model in UAT for performance validation.  
<font color = 'green'>*Note: UAT and Pre-prod can be used interchangeably*. </font>

- Please go to the OpenScale Dashboard, there you will see `Add to Dashboard`/ `Add` button, on the top right corner, that can be used to add a specific deployment for monitoring purpose. 
- This will bring up a popup window to select your UAT machine learning provider, that is, Airline_UAT_MLP_`GroupNumber` which will show a list of available deployments. This list must show the deployment(s) you have created in previous steps.

* Select the appropriate pre-production model deployment and press `Configure`.

On the next screen another dialog box will appear to `Configure Monitors`. Click on the `Configure Monitor` and that will take you to the screen below.



- Edit `Model input` by click on `pencil` icon and provide the following:

   1. Identify Type of Inputs (the type of data used by model) - specify Numerical/Categorical

   2. Identify Algorithm Type - specify Binary Classification
- Press `Save and continue` 


<br>


#### Provide `Training data`
- Click on the `Pencil` icon next to the `Training Data` tab
- Select `Restricted (statistics only)` as the storage type
- Upload the `training_stats.json` file that you downloaded on your system locally while working on the `Create Training Data Statistics JSON` notebook
- Click on `Save` to connect to the training data



<br> 

#### Select `Automatic logging` in the next step. 

+ In this step after you click on `Check now` You will be asked to send initial scoring request with this message. `Scoring request required: Send an initial scoring request to the model and try again`



+ In order to send the scoring request, you can either send payload scoring from the Rest API endpoint or you can use the UI after clicking on the deployment of the model as shown below.


Go to the Model_Deployment_`GroupNumber` deployment in the Pre-prod_Deployment_`GroupNumber` deployment space to follow the UI approach

- You can use the example payload given in the following cell.

<code>
{
	"input_data":[{
	"fields": ["DAY","DAY_OF_WEEK","ORIGIN_AIRPORT","DESTINATION_AIRPORT","DEPARTURE_DELAY","TAXI_OUT","DISTANCE"],
	"values": [[2,5,"PHX","MSP",-2.0,13.0,983]]
	}]
}
</code>

- Go back and click on `Check now`. The error message should not appear now.
- Next, Tick the `Prediction` and `Probability` check boxes and click on `Save`
- Model Details portion has been configured now.
