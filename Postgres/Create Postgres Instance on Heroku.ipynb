# Create Postgres Instance on Heroku 

### Step 1: Sign Up for a Free Heroku Account
Sign up for an account using the following link: https://signup.heroku.com/login 

### Step 2: Create an App
Login to Heroku and select `Create new app` from the homepage: 

Give the app a name and select `Create App`.

### Step 3: Add a Postgres Instance to the App

From the app homepage, navigate to the `Resources` tab:

In the `Add-ons` search bar, search for `postgres` and choose `Heroku Postgres`.  In the next window, ensure the plan name is `Hobby Deb - Free` and then choose `Submit Order Form`.

Open the database by clicking on `Heroku Postgres`:

### Step 4: Get the Database Credentials
From the postgres homepage, navigate to the `Settings` tab and then select `View Credentials` to the right of `Database Credentials`.  Save these credentials for later.

### Step 5: Load Data into the Database

Data can be uploaded into the DB using `psql`.  To start, open up a command line session on your local machine.  Run the following command, using the hostname, database name, and username from the database credentials in the previous step:

`psql -h <hostname> -d <database_name> -U <user_name>`


You will be prompted to enter your password.  Use the password from the db credentials.  Next, create a table in the database using the following command:

`CREATE TABLE IF NOT EXISTS airport_information (FLIGHT_ID INT, ORIGIN_AIRPORT VARCHAR(3), DESTINATION_AIRPORT VARCHAR(3));`

Now insert data into the created table using the following command.  This command assumes the data is in a local csv.

`\copy AIRPORT_INFORMATION FROM '/Users/chrislocandro/Desktop/jan_2015_airports_10000.csv' with (format csv,header true, delimiter ',');`

Verify data was uploaded by navigating to the homepage of the database (confirm number of rows added):

To view the data you can navigate to the `Dataclips` tab and run a `SELECT * ...` query in the SQL editor.
