## Investigate Monitoring Results

#### This Notebook is to show how to access Monitoring results from Watson Open Scale using UI

- For investigating monitoring results of a model, please go to the OpenScale Dashboad.

- Click on the tile of the model and it will bring you to an evaluation summary screen, similar to the following picture.

- You can click on any of the tiles for the different individual monitors (Drift, Fairness or Quality) in order to inspect them in detail.

- Please note that the actual monitoring results may vary based on the algorithm type/parameters. 

### View Fairness

- Please click on Fairness (clickable link under `Fairness`). It will take you to the Fairness screen.
- On the left pane, select one of the attributes you are monitoring for Fairness (`DAY_OF_WEEK` is the example in the picture below). 
- In the middle of the screen, you can specify the values for `Time frame` and `Date range`, or leave the default values. In the example below, `Hourly` and `Today` are chosen, respectively.
- There you can click on a particular point on the line, which will take you to a new screen.

- Click on one of the blue points at the top of the chart (for the 3-3 Monitored Group), and it will take you to the details of the Fairness.

- Now click on the `View Payload Transactions` button on the top right, and see transaction details as shown in the following picture.

### Explainability

- Now, you can click on the `Explain prediction` link for any transactions. That would generate explanation for that transaction. It may take few minutes for Watson OpenScale to generate the Explanation. As shown in the results below, we see the most important factors contributing to such a prediction.


- Additionally, you can click on the `Inspect` tab to analyze controllable features, by scoring with new values in the UI and analyzing the predicted outcome as depicted below. In the example below, the values for `ORIGIN_AIRPORT` and `DESTINATION_AIRPORT` are modified, as indicated by the blue dots. 

- After modifying the values for various features, select `Score new values` and `Run analysis` to view the value for a different outcome.  

<br>

### Quality

- Now go back to the OpenScale Dashboard and click on the tile of your model again. Click on `Quality/Area Under RoC`. It will show you the Quality with respect to Threshold you have set in the configuration step previously. In the middle of the screen, you can specify the values for `Time frame` and `Date range`, or leave the default values. In the example below, `Hourly` and `Today` are chosen, respectively.


- If you click on the line of Quality (green dot in the above picture), you can see the details of the Quality.

<font color = 'red'>Note: For quality feedback, you may upload the same feedback data multiple times at interval of 10-15 minutes for up to 10 times</font>


### Drift
- Now go back to the OpenScale Dashboard and click on the tile of your model again and click on Drift (clickable link under Drift). It will show you the drop in accuracy with respect to Threshold you set in configuration step previously. 

<i>Note: It may show a drop in data consistency over a period of time which can happen due to change in the type of data sent as scoring request.</i>
