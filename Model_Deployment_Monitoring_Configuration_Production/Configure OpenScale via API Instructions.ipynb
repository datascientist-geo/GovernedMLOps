# Create OpenScale Subscription via API


## Overview
This notebook illustrates how to configure a deployment (prod or pre-prod) for monitoring in OpenScale using the OpenScale API.  To do so, you will be using the jobs framework in Cloud Pak for Data.

## Prerequisites

1. WML deployment exists (either your pre-production model deployment or your production wrapper function deployment)
2. Training statisics JSON and drift model file have been generated and are both uploaded to the project as data assets.
3. Sample scoring request JSON and OpenScale configuration notebook are both uploaded to the project. 

## Steps

### 1. Locate the JSON file 'os_template.json' in your CPD project and download it locally.  Open in a local text editor and fill in each of the required parameters as described below:
    - "training_stats_file" : name of training stats JSON that you uploaded to the project
    - "drift_file": name of drift tar.gz file that you uploaded to the project 
    - "cpd_url" : cpd host URL
    - "cpd_username" : Cloud Pak for Data username
    - "cpd_api_key" : Cloud Pak for Data user API Key
    - "deployment_space": Name of WML deployment space where deployment exists
    - "deployment_name": Name of deployment to be monitored in OpenScale
    - "problem_type": "binary"
    - "scoring_json": "sample_scoring_request.json"
    - "quality_metric": "accuracy"
    - "quality_min_records": 50
    - "quality_threshold": 0.96
    - "drift_min_records": 50
    - "drift_threshold": 0.95
    - "op_url": OpenPages instance url 
    - "op_username": OpenPages username (e.g. "ModelManager")
    - "op_password": OpenPages password (e.g. "ModelManager")
    - "op_model_id": OpenPages model name (e.g. "MOD_0000027")

### 3.  Locate the notebook called 'OpenScale Configuration via API' in your project.  Select the pencil icon to open the notebook in edit mode:


### 4. After opening the notebook in edit mode, select the 'File' menu and then select 'Save Version':


### 5.  Select the icon in the top righthand corner with a window and play button.  Then select 'Create a job'.


### 6. In the next page, provide a name for the job e.g. 'Configure OpenScale via API'.  Then select 'Next' in the bottom righthand corner.


### 7. In the Configure section, leave the default values for 'Select version' ('latest') and 'Select environment' ('Default Python 3.7'). Then select the 'Advanced configuration' checkbox below the environment selection. In the box that appears, enter the following: CONFIG_FILE=< _config_file_name.json_> . Use the exact name of the JSON file that you uploaded to the project in Step 2. Then select 'Next' in the bottom righthand corner.


### 8. Leave the schedule screen as is ('Schedule Off'). Then select 'Next'.


### 9. Select 'Create and Run' in the bottom righthand corner. This will trigger the OpenScale subscription to be created using the inputs that you provided. Wait a couple of minutes for the job to complete.


### 10. Navigate to the OpenScale Instance for your Cloud Pak for Data cluster to verify that the subscription was created for your deployment.  The OpenScale Instance URL will be the main CPD host URL with '/aiopenscale/insights' appended to the end, e.g. https://zen-cpd-zen.ibmcloud-roks-sh4fdyvr-6ccd7f378ae819553d37d5f2ee142bd6-0000.eu-gb.containers.appdomain.cloud/aiopenscale/insights
