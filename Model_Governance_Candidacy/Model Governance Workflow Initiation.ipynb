## Model Governance Workflow Initiation

#### In this step, we will use IBM OpenPages `Workflow` feature for Model Governance.

#### We will be using following workflow for Model Governance


#### The various steps we are going to perform as part of `Model Governance Initiation` are creating model object in OpenPages, starting the model workflow, capturing the model details etc. 

### PreRequisites

1. You must have a valid OpenPages URL with the format:
https://openpages-openpages-1-xxxx-xxxx-xxxx/
2. OpenPages credentials to login (e.g. Model Owner)

### Create Model Object

####  - The Model Owner will create `Model` object in OpenPages which will be used for `Model Governance`  throughout the Model Lifecycle right from `Capturing Mode Details` to `Productionizing Model`

#### 1. Login to OpenPages as `Model Owner` <br/>


#### 2. Click the `hamburger` icon on the top right corner <br />



#### 3. Click on the `Inventory` and select `Models` option <br />


#### 4. Click on the `Add New` option to create a new Model object <br />


#### 5. Provide necessary details about the new Model object. All the required fields are marked by `*`. Please make sure 
#### - All the `fields marked with *` must be answered 
#### - The `Save` option will be enabled only after answering all the required fields
#### - The field `Machine Learning Model` should be set to `Yes`
#### - The field `Monitored with Watson OpenScale` will be visible only after the field `Machine Learning Model` is set to `Yes` 
#### - The field `Monitored with Watson OpenScale` should be set to `Yes` <br/>


#### - Scroll down to the `Associations` section and select appropriate Parent Entity for  Model<br />


#### 6.  Click on the `Save` option once the fields are populated with the appropriate values <br/>


### Start Model Governance Workflow

#### 1. On the Model object `Details` page, Click the `Action` on the top right corner and select `Start MRM Workflow`. 


#### 2. A pop-up will be displayed to confirm the start of `Model Governance` workflow. Click on the `Continue` option. This will start the `Model Governance Workflow` on the newly created `Model` object  <br />


### Capture Model Details

#### 1. Once the `Governace Workflow`  is started on the `Model` object. The first step is to `Capture Model Details`.  The Workflow will use a custom view to divide and display information in various parts such as:
#### - The header will display important fields such as `Model Life Cycle Stage`, `Assignees` etc. which user can track throughout the workflow
#### - The section `Capture Model Details` will show fields specific to Capturing Model Details which will help user to focus on the relevant fields
#### - The `Guidelines` section on the right side will help user to understand what he is supposed to do in the specific workflow stage
#### - The `Key fields` section below the `Guidelines` section will help user to focus on the important fields for the specific workflow stage
#### - The `Inline Guidance` (e.g. How to generate Training Data Statistics) will provide user with the necessary guidance while dealing with the fields in the specific section 



#### 2. Capture the Model Details and click on the `Save` option <br/>

#### - Click on Pencil next to Model Catalog and Select the Model catalog allocated for each group. For example, for group1, model catalog should be tagged as `Model_Catalog_Grp1` 
#### - Select the appropriate Model Catalog link by going to Cloud Pak for Data, opening your respective catalog from the left pane and copy the url in the browser when you are in your catalog. 
#### - Add a suitable Model DataSource Description as shown below. <br />



#### 3. Once the Model details are captured, click on the `Action` on the top right corner and select `Validate Model Candidacy` <br />


#### 4. The Workflow will show a pop-up to confirm the `Validate Model Candidacy` action. Click on `Continue and close tab`. <br />


#### 5. Click on `user` icon at the top-right corner and select the `Log Out` option. <br />


#### 6. This concludes the `Model Governance Initiation` module. The next module ( i.e. action ) is `Model Candidacy Validation` which will be performed by `Model Approver`.
