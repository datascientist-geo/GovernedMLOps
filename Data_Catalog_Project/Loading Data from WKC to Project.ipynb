## Loading data from Watson Knowledge Catalog to Watson Studio Project

## Overview and Prerequisites

This notebook is a continuation of "Virtualizing Data using Data Virtualization" notebook. We will be importing the virtualized dataset, "FLIGHT_INFORMATION_FULL" into a Watson Studio Project.

## Prerequisites
1. This notebook assumes you have completed the two prior notebooks, "Adding Postgres and DB2 Connections to CP4D" and "Virtualizing Data using Data Virtualization". 

## Loading data to the project

### 1. Starting from the CP4D home screen, navigate to the Sandwich icon at the top left corner, then go to 'Catalogs' > 'All catalogs'. This will take you to a view of your available data catalogs. Click on 'Default Catalog'. 




### 2. Click on the 'ADMIN.FLIGHT_INFORMATION_FULL' dataset. Then click "Add to Project +" on the top right corner.This will take you to the "Add to project" page
    FYI: The schema is "ADMIN" because we used the ADMIN role to create the virtualized table and left all default settings.






### 3. On the "Add to project" page, select the Airlines-MLops-`GroupNumber` project where you want to add the dataset  and then click on "Add" at the bottom right corner. 

##### In this case,  "MLOps-CPD3.5-..." project is used as an example



### 4. Navigate to the Sandwich icon at the top left corner, then go to 'Projects' > 'All projects'. This will take you to a view of your available Watson Studio projects. 




### 5. Click on the Airlines-MLops-`GroupNumber` project where you added the dataset. You will be able to see the virtualized dataset under the "Data Assets" section. 

##### In this case,  "MLOps-CPD3.5-..." project is used as an example
