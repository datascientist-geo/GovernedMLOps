## Model Monitoring in Production Environment



- Go to the OpenScale Dashboard and click on the tile (see picture below) for the prod deployment you created in the previous step. 

- Now you can click on `Actions`-> `Evaluate now`-> `Evaluate now` option to evaluate the model as shown in the picture below. 
   - Please note that it may take few minutes to few hours before you start seeing the result. You have to press `Evaluate now` again after some time to see the results.

### Upload feedback data for Quality
 
 Feedback data (`test_flights_jan_2015.csv`) is provided as a data asset. Please download it in your system.  

- Click on `Actions`-> `Evaluate now`> `Upload feedback data`. Browse to the feedback file. 

- If it asks for field delimiter, please select `Comma (,)` as shown in the picture below. 

Finally, evaluate by clicking on `Evaluate now`.
