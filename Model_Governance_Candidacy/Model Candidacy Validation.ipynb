## Model Candidacy Validation
#### In this Module, `Model Approver` will validate the Model Candidacy and `Approve\Reject` it based on the details provided by `Model Owner` as part of `Model Governance Workflow Initiation`

### Prerequisites 

1. You must have a valid OpenPages URL with the format:
https://openpages-openpages-1-xxxx-xxxx-xxxx/
2. OpenPages credentials to login (e.g. Model Approver)
3. Captured the Model details as part of `Model Governance Workflow Initiation`

#### 1. Login as `Model Approver`


#### 2. On the `Home Page`, select the `My Tasks` and click on `Model` object ( i.e. Mod_0000003). This will take user to the `Detail` page of the `Model` object. <br/>


#### 3. On the `Model` detail page, click on `Capture Model Details` section to expand it and view the details captured by `Model Owner`. User can use the `Guidelines` provided on the right side to understand the possible actions that they can take on the workflow. Click on the `more` option to view the complete `Guidelines`.  <br/>

#### User can perform one of the following action listed in `Action` button in the top right corner: 
#### - `Reject` the Model Candidacy if the `Captured Model Details` are NOT as expected. The workflow will go back to the previous stage `Capture Model Details` and assign it to previous owner `Model Owner`
#### - `Approve` the Model Candidacy if the `Captured Model Details` are as expected. The workflow will move to the next stage `Data Sourcing` <br />

> #### Note: Please refer `Guidelines for validating Model candidacy` for more details <br />



#### 4. If the `Captured Model Details` are as expected, click on the `Key Items` and provide necessary information as suggested by the `Guidelines`. Click on `Save` option present at the top right corner. <br />


#### 5. Click the `Action` button on the top right corner and select `Data Aquisition` <br />


#### 6. The confirmation window will appear to confirm `Model Candidacy Approval`. Click on `Continue and close tab`. This will move the `Model Governance Workflow` to the `Data Sourcing` stage which will be performed by `Data Engineer` <br/>


#### 7. Click on `user` icon at the top-right corner and select the `Log Out` option. 


#### 8. This concludes the `Model Candidacy Validation` module. The next module ( i.e. action ) is `Data Sourcing` which will be performed by `Data Engineer`
