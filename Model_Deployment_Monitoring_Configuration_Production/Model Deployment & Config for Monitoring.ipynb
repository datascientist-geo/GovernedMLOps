## Model Deployment and Configuration for Monitoring  in Production Environment


#### This step involves exporting the desired model which was deployed in UAT environment and was later approved to move from from UAT environment to production.  

### Step 1. UAT Deployment Space 

Go to the Navigation menu of your Cloud Pak for Data Platform, and select `Deployments` to view all your deployment spaces.

Open the "Prod_Deployment-`GroupNumber`" deployment space

#### Go to `Access Control` of this deployment space and make sure that there is a user named `admin` in the Collaborators section. 
- This is needed for Openscale

<i><b>Note:</b>  If you are logged in as ‘admin’, then you would not have to check for ‘admin’ as a collaborator. </i>
 

<br>

### Step 2. Model Promotion in Production Space

Follow the steps below to deploy the model in your Production Deployment Space. 
   
   - Go to the `Assets` tab of your 'Airlines-MLops-`GroupNumber` project. 
   - Scroll down to the list of the Models. 
   - Select the model with the best performance metrics
   - Go to the model you want to publish to the "Prod_Deployment_`GroupNumber`"
   - Hover on the triple vertical dots at the right of the model name and click `Promote` as given below



If the "Prod_Deployment_`GroupNumber`" space is associated to the project, You will be on the "Promote to Space" page. 

Make sure that the target space is "Prod_Deployment_`GroupNumber`". Click on `Promote` and the model will be promoted to that deployment space successfully.

<br>

<i> 
    
Note: If a space is not associated with the project yet, it would popup an option to associate a deployment space. 
    
- In the popup, select `Associate Deployment Space`. 
- In the next window, click on `Existing` tab
- Select the "Pre_Prod_Deployment_`GroupNumber`" UAT space then click `Associate`. 
- Now, promote the model again. This time the model should be promoted to the deployment space.
    
</i> 

<br><br>
+ After that is done, you will be able to see the imported model in the production deployment space as given in the screenshot below.


<br><br>
### Step 3. Prod Model Deployment

+ The Model will appear in the `Assets` tab . To deploy your desired model, click on the `Deploy` icon as shown in the image below.


You will now reach the "Create a Deployment" page.  

- Select `Online` as the deployment type
- Give a name to your model deployment, for example, "Prod_Model_Deployment_`GroupNumber`"
- Add additional description, if required
- Finally, click on the `Create` button.

<br>

You will now be back in your "Pre_Prod_Deployment_`GroupNumber`" deployment space

- Go to the `Deployments` tab where you will be able to see the status of your model deployment.
- The deployment status will change from `In progress` to `Deployed` once the model deployment is successful completed.
      
<br>

<br><br>
### Step 4. Configuring Prod Model in OpenScale


#### Add Model for Monitoring

- Go to the OpenScale dashboard. Click on `Add to Dashboard` or `Add`. 
- Select the production Machine learning provider, that is, Airline_Prod_ML_Provider_Grp1
- It will show a list of deployments in a popup window.
- Select your prod deployment and press `Configure`. 
    

- Press `Configure monitors` in the next dialog box.


- And Click on `Import settings` (see picture below).

- In the next window (see picture below), a list of deployments will appear. Select the deployment you configured for UAT and press `Configure`.

 + Please wait for sometime. It imports the settings of the model you configured for UAT/pre-prod environment. 


- When the import is done, all the parameters (Fairness/Quality/Model parameters etc.) and configuration/settings will automatically be imported.
