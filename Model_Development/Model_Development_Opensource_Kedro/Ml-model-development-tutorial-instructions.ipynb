# ML Model Development Tutorial

1. First, download the **mlops-model-development-opensource.zip** from the github repo.

2. Create a project with Github integration. The instructions can be found in <mark>Github-integration-instructions</mark> notebook.

3. Once the project is created, click on the assets section. Then load the data and the "mlops-model-development-opensource.zip" files via `find and add data button` on the right top corner (arrow in the below image). Either drop the files or click the browse button  (red rectangle) to add files to the project.

4. Next, click on the "Launch IDE" button shown with arrow in below image and click on JupyterLab to start Jupyter Lab session.

5. Select GovMLOps-JupyterLab for the environment and select your Git token from the list. Then click Start button. Note that the instructor may ask you to create a custom environment based on the usecase. 

6. The Launcher window will appear once the platform configures the JupyterLab session. From the window, select Terminal icon listed under Other section (red arrow in below image).

7. In the terminal window, check whether kedro is installed by executing <code> kedro info </code> command. The command should show kedro graffiti and version number as shown below. 

<p> If your command does not return above image or shows "command not found" error, then please install the kedro opensource package by running <code>pip install kedro==0.17.0</code> 

Once the installation is done, please run <code>kedro info</code> again to verify whether the installation was successful (The kedro graffiti and version number should appear)</p>

8. Next, unzip the **mlops-model-development-opensource.zip** zip file. First, verify the file is located under <mark>project_data_assets/data_asset/</mark> folder.

Run <code>unzip project_data_assets/data_asset/mlops-model-development-opensource.zip</code> This will extract the content of zip file and create a folder called *airline-kedro*. 

9. Change the directory to airline-kedro <code> cd airline-kedro </code> in the terminal window.
<p>Then change the directory on the left pane (shown in the red rectangle) and navigate into the <mark>airline-kedro</mark> folder as well.

10. From the left pane, navigate to the catalog.yaml file from double clicking <mark> conf -> base -> catalog.yml </mark> Then change the file path where you add your data. Once you are done with editing the catalog.yml file, you can save your changes by clicking <mark> File -> Save YAML file </mark>.

11. From the left pane, click on the ... shown in the red rectangle to navigate back to the main airline-kedro folder. Then, double click on <mark>src -> airline_kedro -> pipelines</mark> directories.

12. Inside the <mark>pipelines</mark> directory, double click on the model development folder and double click on the **nodes.py** python script. Scroll down to the save_to_catalog function, change the *zip_file_name* and *name_of_catalog* variables. Then save the python script via clicking on the save button shown in red square. 

13. From your terminal window, make sure you are in */home/wsuser/work/airline-kedro* folder running <code>pwd</code> command. Then, simply run the pipeline using <code>kedro run</code> command. 

14. Once the command is executed, return back to your project page and verify you have the model in the Models asset section. In addition, go to the catalogs and selecy your catalog. Finally, verify your model and zip package are pushed to the catalog.
