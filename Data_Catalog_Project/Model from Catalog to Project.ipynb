## Adding model from Model Catalog to "Airlines-MLOps-Prod-`GroupNumber`" project

#### In the previous notebook, Developer pushed the developed model to catalog. Validator would not have access to the Airlines-MLops-`GroupNumber` project. Now, the validator must login to the catalog, get the model and push it to Airlines-MLops-Prod-`Groupnumber` project.

#### This notebook involves steps to add the model from "Model-Catalog-`GroupNumber`" Catalog to "Airlines-MLOps-Prod-`GroupNumber`" Project

1. Go to the Navigation menu on your Cloud Pak for Data Platform. From there, Select Catalogs->All Catalogs



2. Open the "Model-Catalog-`GroupNumber`" Catalog
3. Select the model you would like to add to the "Airlines-MLOps-Prod-`GroupNumber`" project, Click on the `three vertical dots` icon next to it and select `Add to Project`




4. Now, select the  "Airlines-MLOps-Prod-`GroupNumber`" project and click on `Add`



<br> 

#### The validator can now use the model from the "Airlines-MLOps-Prod-`GroupNumber`" project for further UAT/Pre-prod staging
