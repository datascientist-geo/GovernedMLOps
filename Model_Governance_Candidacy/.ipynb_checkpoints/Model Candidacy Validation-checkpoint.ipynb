## Model Candidacy Validation

### Prerequisites 
#### - Capture the Model details as part of `Model Governance Workflow Initiation`

#### 1. Once the Model details are captured, click on the `Action` on the top right corner and select `Validate Model Candidacy` <br />


#### 2. The Workflow will show a pop-up to confirm the `Validate Model Candidacy` action <br />


#### 3. Login as `Model Approver`

#### 4. Go through the `Captured Model Details` and perform one of the following action listed in `Action` button in the top right corner: 
#### - `Reject` the Model Candidacy if the `Captured Model Details` are NOT as expected. The workflow will go back to the previous stage `Capture Model Details` and assign it to previous owner `Model Owner`
#### - `Approve` the Model Candidacy and proceed for `Data Sourcing` if the `Captured Model Details` are as expected. The workflow will move to the next stage `Data Sourcing` <br />

> #### Note: Please refer `Guidelines for validating Model candidacy` for more details <br />





#### 4. If the `Captured Model Details` are as expected, click the `Action` on the top right corner and select `Data Aquisition` <br />


#### 5. This will take user to the `Data Sourcing` section
