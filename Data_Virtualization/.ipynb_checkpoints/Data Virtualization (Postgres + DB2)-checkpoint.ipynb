## Virtualizing Data Using Data Virtualization

## Overview and Prerequisites

This notebook is a continuation of "Db2 and Postgres Connection.ipynb". We will be virtualizing and merging 2 databases, one on Db2 on CP4D (Flight Information), another on Postgres (Airport Information).  

## Prerequisites
1. This notebook assumes that the Flight Information and Airport Information datasets are loaded into IBM CP4D and Postgres respectively.
2. This notebook assumes you have created Db2 data connections on CP4D, FlightInfoCP4D and Postgres DB, for IBM CP4D and Postgres respectively. 
3. This notebook assumes that you have an enabled instance of IBM Data Virtualization and Watson Knowledge Catalog on Cloud Pak for Data (CP4D).

## Virtualizating Data Using Data Virtualization

#### 1. Navigate to the Sandwhich icon at the top left corner, then go to 'Data' > 'Data Virtualization'. This will take you to a view of your data sources available for virtualization.



#### 2. Navigate to 'Add data source'>'Select existing connection'. This will take you to the page with existing connections (these were created in the prior notebook). 



#### 3. Click 'FlightInfoCP4D'> 'Add'. Skip the optional 'Add to a remote connector'. This will add your connection to Data Virtualization.



#### 4. Reapeat steps 2 and 3 for 'Postgres DB'. Once these steps are completed, you should see the connections on the 'Data sources' page.



#### 5.  Navigate to 'Data sources at the top left, select the dropdown, and navigate to 'Virtualize. This will take you to the page listing all tables eligble for virtualization. 



#### 6. Type 'information' in the searchbar on the 'Virtualize' page. You will be able to see the 'FLIGHT_INFORMATION' (located in CP4D) and 'AIRPORT_INFORMATION' (located in Postgres) tables. Check the tables and click 'Add to cart' >'View Cart'.

#### 7. Click on 'View Cart'. This will take you to the 'Review cart and virtualize tables' page.

#### 8. Select a project to where you want to add the virtualized data. Leave the other default settings and click 'Virtualize' at the top right corner. 

#### 9. Once the data is virtualized, a 'Virtualize Objects' screen will pop up. Click on 'View my virtualized data'. This will take you to the 'My virtualized data' page. 


#### 10. In the 'My virtualized data' page, check the boxes on FLIGHT_INFORMATION and AIRPORT_INFORMATION (now virtualized datasets) and click 'Join'. This will take you to the 'Join virtual objects' page. 


#### 11. On the 'Join virtual objects' page, click and drag 'FLIGHT_ID' from the left hand table (FLIGHT_INFORMATION) to 'FLIGHT_ID on the right hand table (AIRPORT_INFORMATION). Click 'Preview' and then 'Next'. 
    FYI: You defined the merge key in this step. 
    
   

#### 12. Click 'Next' on the 'Edit column names' page. This will take you to 'Assign and review'. Name the merged table 'FLIGHT_INFORMATION_FULL', select the project for the joined table, and click 'Create view'.





#### 13. You should now see the merged table 'FLIGHT_INFORMATION_FULL' in the 'My virtualized data' screen. Check the box next to the table name and click 'Publish'. 



#### 14. Navigate to the Sandwhich icon at the top left corner, then go to 'Catalogs' > 'All catalogs'. This will take you to a view of your available data catalogs. Click on 'Default Catalog'. You will be able to see the 'FLIGHT_INFORMATION_FULL' dataset.
