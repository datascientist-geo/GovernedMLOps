# Push Model from Pre-Prod to Prod Deployment Space (CICD)

## Overview
In this notebook you will push a model from a pre-prod deployment space to a prod deployment space as part of a CICD workflow. 

## Prerequisites
1. Pre-prod model has been deployed in a pre-prod deployment space.
2. Prod deployment space exists.

## Steps

### 1. The notebook which will be used to push the model from pre-prod to prod has already been uploaded to your CPD project - it is called *Move Model from Pre-Prod to Prod*. Locate this notebook and open it in Edit mode:


### 2. Once the notebook is opened in edit mode, select the 'File' menu and then select 'Save Version'.


### 3. Select the icon with a window and a play button in the top righthand corner.  Then select 'Create a job'


### 4. Give the job a name ('Move Model from Pre-Prod to Prod Job'):


### 5a. Leave the default values for 'Select version' ('Latest') and 'Select environment' ('Default Python 3.7').  Then select the 'Advanced Configuration' checkbox.  In the resulting box, enter the following: 

PRE_PROD_SPACE_NAME=Pre-Prod Deployment_Grp*INSERT YOUR GROUP NUMBER* <br>
PRE_PROD_DEPLOYMENT_NAME=*INSERT NAME OF YOUR PRE PROD DEPLOYMENT* <br>
PROD_SPACE_NAME=Prod Deployment_Grp*INSERT GROUP NUMBER* <br>
PROD_DEPLOYMENT_NAME=Prod Flight Model Deployment Grp*INSERT YOUR GROUP NUMBER* <br>
    
See below screencap for an example: 

### 5b. Select 'Next' in the bottom righthand corner (after entering Advanced configuraiton variables)

### 6.  Leave the Schedule *as is* and then select 'Next' in the bottom righthand corner.


### 7. In the 'Create a job' page, select 'Create and Run' in the bottom righthand corner:


### 8. This will trigger the job to run and the model will be pushed to the prod space.  Wait for the job to complete and then navigate to the prod deployment space and verify that the deployment was created successfully.
