# Creating a new project and adding assets in your CPD platform
This notebook will show you how to create new empty projects and add project assets like notebooks, datasets, etc in your CPD platform 

### Prerequisites 
Login to your Cloud Pak for Data Platform

### Creating the project

#### Step 1 - Go to the sandwich menu at top left corner of the window, Select on Projects. This will take you to the Project's page

#### Step 2 - Click on the 'New project +' button as shown in the image below.

#### Step 3 - Select <mark>Analytics project</mark> as the project type and click on Next button

#### Step 4 - Select Create an empty project button as shown with red arrow below image

<mark> Note - If you would like to integrate this project with github then go to the `Git Integration Instructions` notebook, else continue with this notebook </mark>

#### Step 4 - Fill in the define project details. 

- Enter name of the project in Airline-MLOPs-`Groupnumber` format and click on `Create` tab at the bottom right corner.


### Adding assets 

For this use case we will add a python notebook to the project, you can add other assets like datasets, connections, models etc.

#### Step 1 - Go to the `Assets` section of your newly created project, click on `Add to Project`, and select `Notebook`


#### Step 2 - Go to `From File`, select your desired runtime, we have selected Python 3.7 in this case and then drag files from your local computer in the drag and drop section. Click on `Create`. 


#### You will be able to see your notebook in the `Notebooks` section of the project assets. 

<mark> Please make sure to properly exit your notebook by stoping the kernel as given in the below screenshot </mark>

#### Unlock the notebook, go to 'vertical dots' icon of the respective notebook and then click on `Stop Kernel` to avoid unnecessary utilization of resources.
